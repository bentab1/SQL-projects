```python
import numpy as np
import pandas as pd

# Load your orders file
orders = pd.read_csv("orders.csv")

# Add a 'paid' column (95% paid, 5% unpaid)
orders["paid"] = np.random.choice([True, False], 
                                  size=len(orders), 
                                  p=[0.95, 0.05])

# Add a 'cart_items' column assuming e-commerce platforms store item count
# Let's simulate between 1 and 8 items per order
orders["cart_items"] = np.random.randint(1, 9, size=len(orders))

# Save updated file
orders.to_csv("orders_with_paid_and_cart_items.csv", index=False)
```
